## gc_aangle

`gc_aangle` calculates acute angle between two great circles on the globe

### Grab and Go

```python
lat = [10.0, 20.0, 0.0, 0.0]
lon = [0.0, 0.0, 5.0, 20.0]

arc1 = [(lat[0], lon[0]), (lat[1], lon[1])]
arc2 = [(lat[2], lon[2]), (lat[3], lon[3])]
print(arc1)
print(arc2)
```

## gc_clkwise

`gc_clkwise` determines the clockwise/counterclockwise ordering of points on spherical polygon

## gc_dangle

`gc_dangle` determines the directed angle between two great circles having a specified intersection point

## gc_inout

`gc_inout`  determines if a list of latitude/longitude specified points are inside or outside of spherical latitude/longitude polygons

## gc_latlon

`gc_latlon`  determines ithe great circle distance (true surface distance) between two points on the globe and interpolates points along the great circle

## gc_onarc

`gc_onarc`  determines if a point on the globe lies on a specified great circle arc

## gc_pnt2gc

`gc_pnt2gc`  determines the angular distance from a point to a great circle

## gc_qarea

`gc_qarea` determines the area of a quadrilateral patch on the unit sphere

## gc_qarea

`gc_qarea` determines the area of a triangular patch on the unit sphere
